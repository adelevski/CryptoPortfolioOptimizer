```python
coins = ['AAVE','ADA','ALGO','ANKR','ANT','ATOM1','BAL','BAND','BAT','BCH','BNT','BTC','CHZ','COMP','CRV','DASH','DOGE',
 'DOT1','ENJ','EOS','ETC','ETH','EWT','FIL','GNO','GRT2','ICX','KAVA','KNC','KSM','LINK','LRC','LSK','LTC','MANA',
 'MATIC','MIR1','MKR','MLN','NANO','OMG','OXT','QTUM','REP','SAND','SC','SNX','SOL1','SRM','STORJ','SUSHI',
 'TRX','UNI3','WAVES','XLM','XMR','XRP','XTZ','YFI','ZEC','ZRX']

for i in range(len(coins)):
    coins[i] = coins[i] + '-USD'
```

```python
end_date = dt.datetime.now()
start_date = end_date - dt.timedelta(days=3)
```

```python
mean_returns, cov_matrix = ef.getData(coins, start_date, end_date)
```

```python
maxSR_returns, maxSR_std, maxSR_allocation, \
minVol_returns, minVol_std, minVol_allocation, \
efficientList, targetReturns = ef.calculatedResults(mean_returns, cov_matrix)
```

```python
maxSR_allocation = maxSR_allocation.loc[maxSR_allocation['allocation'] > 0.0005]
maxSR_allocation.reset_index(inplace=True)
with pd.option_context('display.float_format', '{:.3f}'.format):
    display(maxSR_allocation, maxSR_returns, maxSR_std)
print(maxSR_allocation['allocation'].sum())
```

```python
minVol_allocation = minVol_allocation.loc[minVol_allocation['allocation'] > 0.0005]
minVol_allocation.reset_index(inplace=True)
with pd.option_context('display.float_format', '{:.3f}'.format):
    display(minVol_allocation, minVol_returns, minVol_std)
print(minVol_allocation['allocation'].sum())
```

```python
combined = pd.concat([maxSR_allocation, minVol_allocation]).groupby('index')['allocation'].sum().reset_index()
combined['allocation'] = combined['allocation'] / 2
with pd.option_context('display.float_format', '{:.3f}'.format):
    display(combined)
print(combined['allocation'].sum())
```
